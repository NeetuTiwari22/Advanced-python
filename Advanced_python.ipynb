#View Vs Copy

```python
# View means modification in original data.
# copy means modification in duplicate data that will not reflect original data.
```

```python
a = np.array ([1,2,3,4,5,6,7,8])
a
```

```python
b = a[2:5]
b[:] = 0
b
```

```python
a
```

```python
a = np.array ([1,2,3,4,5,6,7,8])
b = a[2:5].copy()
b[:] = 0
b
```
