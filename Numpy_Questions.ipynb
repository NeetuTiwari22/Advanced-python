```python
import numpy as np
```

```python
# 1
arr1 = np.arange(5, 51, 5)
print(arr1)
```

```python
# 2
arr2 = np.ones((4, 3))
print(arr2)
```

```python
# 3. 6 equally spaced values between 0 and 1.
arr3 = np.linspace(0, 1, 6)
print(arr3)
```

```python
# 4. Create a 3×3 identity matrix.
arr4 = np.eye(3)
print(arr4)
```

```python
# 5. shape, size, number of dimensions, and data type of a given array.
print(arr1.shape)
print(arr1.size)
print(arr1.ndim)
print(arr1.dtype)
```

```python
# 6. the Python list [10, 20, 30, 40, 50] into a NumPy array.
lst = [10, 20, 30, 40, 50]
arr5 = np.array(lst)
print(arr5)
```

```python
# 7. 1D array of 16 elements into a 4×4 matrix.
arr6 = np.arange(1, 17)
matrix = arr6.reshape(4, 4)
print(matrix)
```

```python
# 8. 2D array into 1D using both flatten() and ravel().
flat1 = matrix.flatten()
flat2 = matrix.ravel()

print(flat1)
print(flat2)
```

```python
# 9. Access the middle element of a 1D array.
arr7 = np.array([10, 20, 30, 40, 50])
middle = arr7[len(arr7)//2]
print(middle)
```

```python
# 10. a NumPy array without using loops.
rev = arr7[::-1]
print(rev)
```

```python
# 11.all even-indexed elements
print(arr7[::2])
```

```python
# 12. the last column from a 2D array.
print(matrix[:, -1])
```

```python
# 13. 1D array to get alternate elements starting from index 1.
print(arr7[1::2])
```

```python
# 14. Extract a 2×2 sub-matrix from a 3×3 matrix.
mat3 = np.arange(1, 10).reshape(3, 3)
sub = mat3[:2, :2]
print(sub)
```

```python
# 15. Print all elements except the first and last element of an array.
print(arr7[1:-1])
```

```python
# 16. greater than 25 from a NumPy array.
print(arr1[arr1 > 25])
```

```python
# 17.negative values in an array with 0.
arr8 = np.array([10, -5, 20, -2, 30])
arr8[arr8 < 0] = 0
print(arr8)
```

```python
# 18.Extract only the elements divisible by 3.
print(arr1[arr1 % 3 == 0])
```

```python
# 19.Count how many elements are greater than the mean of the array.
mean_val = arr1.mean()
count = np.sum(arr1 > mean_val)
print(count)
```

```python
# 20.Find the indices of elements where the value is equal to 50.
indices = np.where(arr1 == 50)
print(indices)
```

```python
# 21.Add two NumPy arrays element-wise.
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print(a + b)
```

```python
# 22. Perform matrix multiplication of two compatible matrices.
m1 = np.array([[1, 2], [3, 4]])
m2 = np.array([[5, 6], [7, 8]])
result = np.dot(m1, m2)
print(result)
```

```python
# 23.Add a scalar value 10 to every element of a NumPy array.
print(arr7 + 10)
```

```python
# 24.Add a 1D array of shape (3,) to a 2D array of shape (2, 3) using broadcasting.
arr2d = np.array([[1, 2, 3],
                  [4, 5, 6]])

arr1d = np.array([10, 20, 30])

print(arr2d + arr1d)
```
